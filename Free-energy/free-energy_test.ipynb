```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as sps
```

## Exercise 1

```python
u = 2        # observed input
sigma_u = 1  # observation error
v_p = 3      # expected value
sigma_p = 1  # expectation error
```

```python
x_min = 0
x_max = 5
dv = 0.05
x = np.linspace(x_min, x_max, num=int((x_max-x_min)/dv))
expect = sps.norm.pdf(x, v_p, sigma_p) # p(v)
predict = expect * sps.norm.pdf(x ** 2, u, sigma_u) # p(v|u) = p(v)*p(u|v)
# print(predict)
normalization = sum ( predict * dv )
predict = predict/normalization

with sns.plotting_context("paper", font_scale=1.5), sns.axes_style("whitegrid"):
    sns.lineplot(x=x, y=expect).set(xlabel='$v$', ylabel='$p(v|u)$')
    sns.lineplot(x=x, y=predict).set(xlabel='$v$', ylabel='$p(v|u)$')
    plt.axvline(x=np.sqrt(2), color='orange')
    plt.axvline(x=3)
```
